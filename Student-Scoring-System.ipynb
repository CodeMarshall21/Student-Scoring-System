```python
from sklearn.model_selection import train_test_split
```

```python
feature = dataframe[['Hours_Studied']]
```

```python
dependent_data = dataframe['Marks']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(feature, dependent_data, test_size=0.2, random_state=42)
```

```python
X_train
```

```python
from sklearn.linear_model import LinearRegression
```

```python
linearRegressionModel = LinearRegression() # object
```

```python
linearRegressionModel.fit(feature, dependent_data)
```

```python
linearRegressionModel.predict(X_test)
```

```python
dataframe
```

```python
prediction = linearRegressionModel.predict(X_test)
for Hours_studied, Marks in zip(X_test['Hours_Studied'], prediction):
    print(f"Hours studied: {Hours_studied} -> Marks: {Marks}")
```

```python
manual_hours_input = 10
marks_obtained = linearRegressionModel.predict([[manual_hours_input]])
print(f"Hours Input = {manual_hours_input} -> Marks obtained: {marks_obtained}")
```
