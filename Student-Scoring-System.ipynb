```python
# Visualize data

import matplotlib.pyplot as plt
```

```python
# plot data

x_axis_data = dataframe['Hours_Studied']
y_axis_data = dataframe['Marks']

plt.scatter(x_axis_data, y_axis_data)
```

```python
# label graph
plt.scatter(x_axis_data, y_axis_data)
plt.title("Hours Vs Marks")
plt.xlabel("Hours studied")
plt.ylabel("Marks scored")
plt.show()
```
